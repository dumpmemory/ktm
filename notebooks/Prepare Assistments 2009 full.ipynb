We should check for duplicate lines.

```python
import pandas as pd

df = pd.read_csv('data/assistments2009full/fulldata.csv')
df.head()
```

```python
df['school_id'].nunique()
```

```python
df['list_skills'].value_counts()
```

```python
import numpy as np

df['skill'] = np.unique(df['list_skills'].fillna(''), return_inverse=True)[1]
df['skill'].value_counts()
# This was a fast way; should do it properly with a q-matrix.
```

```python
MAPPING = {
    'user_id': 'user',
    'problem_id': 'item',
    'skill': 'skill',
    'correct': 'correct',
    'school_id': 'school_id'
}
subset = df[MAPPING.keys()].rename(columns=MAPPING).copy()

subset['wins'] = 0
subset['fails'] = 0
subset['correct'] = subset['correct'].astype(int)
subset['school_id'] = subset['school_id'].fillna(0).astype(int)
subset.to_csv('data/assistments2009full/data.csv', index=False)
```
